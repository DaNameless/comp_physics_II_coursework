## 1. Population dynamics (8 points)

The system of ordinary differential equations (ODEs) describing the population dynamics of two prey species ($x$ and $y$) and one predator species ($z$) is given by:


\begin{aligned}
\frac{dx}{dt} &= g_1\,x \left( 1 - \frac{x}{c_1} \right) - p_1\,x\,z, \\
\frac{dy}{dt} &= g_2\,y \left( 1 - \frac{y}{c_2} \right) - p_2\,y\,z, \\
\frac{dz}{dt} &= e_1\,p_1\,x\,z + e_2\,p_2\,y\,z - d\,z.
\end{aligned}


where:

- $g_1$, $g_2$ are the intrinsic growth rates (e.g. birth rates) of the prey populations,

- $c_1$, $c_2$ are the carrying capacities of the prey populations (the carrying capacity of an environment is the maximum population size of a biological species that can be sustained by that specific environment), 
  
- $p_1$, $p_2$ are the predation rates of the predator on each prey (e.g. how successful a hunt is),

- $e_1$, $e_2$ are the conversion efficiencies of consumed prey into predator biomass (the conversion efficiency tells us how efficiently a predator can use the energy from its prey to reproduce),

- $d$ is the natural death rate of the predator. Note that the death rates of the two prey species are intrinsically given by their individual carrying capacities.

This system models the interactions where both prey species grow logistically but are consumed by a common predator.

### State vector and slope:

**(a)** Write down this system of ODEs in terms of the system state vector $S(t)$, i.e. $\frac{dS}{dt} = F(S)$. Identify the slope function $F(S)$ and indicate whether or not it can be explicitly written as a function of $S$. Create a python function for the slope $F(S)$.

We define the state vector as 
$$
\vec{S} =\left(\begin{array}{l}
x \\
y \\
z
\end{array}\right)
$$
where $S_0 = x$, $S_1 = y$, $S_2 = z$ 

And the slope vector function $\vec{F \left( \vec{S} \right)}$ as
$$\vec{F \left( \vec{S} \right) }=\left(\begin{array}{l}
g_1\,x \left( 1 - \frac{x}{c_1} \right) - p_1\,x\,z, \\
g_2\,y \left( 1 - \frac{y}{c_2} \right) - p_2\,y\,z, \\
e_1\,p_1\,x\,z + e_2\,p_2\,y\,z - d\,z.
\end{array}\right)$$

Or, using a matrix (however it may not be the best way to represent the slope vector function)
$$
\vec{F(\vec{S})}  =\left(\begin{array}{ccc}
g_{1}(1-\frac{S_{0}}{c_{1}})-p_{1} S_{2} & 0 & 0 \\
0 & g_{2}(1-\frac{S_{1}}{c_{2}})-p_{2} S_{2} & 0 \\
0 & 0 & e_{1} p_{1} S_{0}+e_{2} p_{2} S_{1}-d
\end{array}\right) \vec{S} 
$$

Notice that **we cannot write** $\vec{F}$ **explicitely as a function of the state vector** $\vec{S}$ as the functions are non-linear. This can be seen in the matrix representation, as we need to call $\vec{S}$ components in various instances.
Also, $\vec{F}$ is not an explicit function of time, therefore we can avoid defining t as a parameter in the slope functions, however, just for completeness we may do it.

```python
def f_slope(pdic, t, s):
    """
    Function representing the slope F(S) of the given system of ODEs 
    Input:
        s (np.array)-> State vector s.t s[0]=x, s[1]=y, s [2]=z.
        pdic (dictionary)-> dictionary containing the parameters in the system, i.e, g1, g2, c1, 
                            c2, p1, p2, e1, e2, d, etc
        t (float) -> Time instance at which we are computing the slope
    Output:
        slope (np.array)-> Slope vector function   
    Author: R.S.S.G.
    Date created: 10/03/2025
    """
    # Get state variables
    x, y, z = s[0], s[1], s[2]

    # Unpack parameters
    g1, g2, c1, c2, p1, p2, e1, e2, d = pdic['g1'], pdic['g2'],pdic['c1'],pdic['c2'],\
                                        pdic['p1'], pdic['p2'], pdic['e1'], pdic['e2'],\
                                        pdic['d']
    # Getting the slope
    slope = np.array([g1*x*(1-(x/c1))-p1*x*z,
                      g2*y*(1-(y/c2))-p2*y*z,
                      e1*p1*x*z + e2*p2*y*z - d*z])
    return slope
```

```python
# Alternatively, we could write it as the following
def f_slope_matrix(pdic, t, s):
    """
    Function representing the slope F(S) of the given system of ODEs, using matrix notation 
    Input:
        s (np.array)-> State vector s.t s[0]=x, s[1]=y, s [2]=z.
        pdic (dictionary)-> dictionary containing the parameters in the system, i.e, g1, g2, c1, 
                            c2, p1, p2, e1, e2, d, etc
        t (float) -> Time instance at which we are computing the slope
    Output:
        m (np.array of arrays)-> Matrix obtained to represent the state vector derivative
        slope (np.array)-> Slope vector (Result of matrix operating of S vector)    
    Author: R.S.S.G.
    Date created: 10/03/2025
    """

    # Unpack parameters
    g1, g2, c1, c2, p1, p2, e1, e2, d = pdic['g1'], pdic['g2'],pdic['c1'],pdic['c2'],\
                                        pdic['p1'], pdic['p2'], pdic['e1'], pdic['e2'],\
                                        pdic['d']

    # Defining matrix
    m = np.array([[g1*(1 - (s[0]/c1))-(p1*s[2]), 0, 0],[0, g2*(1 - (s[1]/c2))-(p2*s[2]), 0],[0, 0, (e1*p1*s[0])+(e2*p2*s[1])-d]])
    # Getting the slope
    slope = np.dot(m,s)
    return m, slope
```

### ODE integration methods:

**(b)** Create a python function that implements a trapezoidal Euler method for ODE integration.


As we have a slope that cannot be explicitly written in terms of the state vector $\vec{S}$, we need a predictor and corrector routine.

*(We may think about adding an iteration routine to repeat the correction step until it reaches a certain tolerance value, however we do not really know if that would help substantially, thus we have just implemented the usual Trapezoidal Method)*

```python
def trpz(f, pdic):
    """
    Function implementing the Trapezoidal Euler method for ODE integration. 
    It uses Euler explicit for the predictor step.
    
    Input:f (python function) -> Function computing the slope F(S). It must be in the form `f(pdic, t, s)`
        pdic (dict) -> dictionary containing the parameters in the system, i.e, g1, g2, c1, 
                            c2, p1, p2, e1, e2, d, x0, y0, z0, t_span, dt, method.
    Output: (np.array of float-containing arrays) -> Array containing the time evolution of the State Vector
    Author: R.S.S.G.
    Date created: 10/03/2025
    """
    # Unpack necessary parameters from dictionary
    t_span = pdic['t_span']
    dt = pdic['dt']
    s0 = np.array([pdic['x0'], pdic['y0'], pdic['z0']])
    
    # Initialize time array
    t_axis = np.arange(t_span[0], t_span[1] + dt, dt)
    n = len(t_axis)
    
   # Initialize solution array and set initial condition
    s = np.zeros((n, len(s0)))
    s[0] = s0
    
    # Time integration
    for j in range(n - 1):
               
        # Predictor step: Explicit Euler
        s_pred = s[j] + dt * np.array(f(pdic, t_axis[j], s[j]))
        # Corrector step: Trapezoidal method
        s_corr = s[j] + (dt / 2) * (np.array(f(pdic, t_axis[j], s[j])) + np.array(f(pdic, t_axis[j],s_pred)))
        # Store the computed state
        s[j+1] = s_corr
    return s
```

Alternatively, we can avoid the predictor and corrector steps (explicitely at least, as we're making the prediction step implicitely in the matrix) by using the matrix representation shown before. Here, we only need to get the appropiate transformed matrix to perform the trapezoidal method.

```python
# Let's implement a function that computes the correct matrix to be used in each Euler method
# The next two functions were written weeks ago to help in the manual processes to obtain
# implementations of each method for systems of ODE's or order reducted systems.
def red_ord_matrix(m, h_step, method):
    """
    This function calculates the matrix obtained from reduction of order and transforms it to be directly applied for 
    Euler Methods. Requires inv from numpy.linalg
    Input: m (np.array of floats) -> Matrix representing the system of equations at reducing the order
            h_step (float) -> step size
            method (str) -> Euler method to use
                            Options: "e" for Explicit, "i" for Implicit, "t" for Trapezoidal
    Output: m_t (np.array of float) -> Transformed matrix obtained from the process
    Author: R.S.S.G.
    Date created: 10/03/2025
    """
    iden = np.identity(np.shape(m)[0])
    if method == "e":
        m_t = iden + h_step*m
    elif method == "i":
        m_t = la.inv(iden- h_step*m)
    elif method == "t":
        m_t = np.dot(la.inv(iden - (h_step/2)*m), iden + (h_step/2)*m)
    return m_t
```

```python
# Let's implement the trapezoidal Euler method using matrices
def trpz_m(f_slope_matrix, pdic, red_ord_matrix):
    """
    This function calculates the evolution of the State Vector using the Trapezoidal Euler method.
    Input: f_slope_matrix (python function) -> Function representing the slope F(S) using matrix 
                                                notation
            pdic (dictionary) -> dictionary containing the parameters in the system, i.e, g1, g2, c1, 
                            c2, p1, p2, e1, e2, d, x0, y0, z0, t_span, dt, method.
            red_ord_matrix (python function) -> Function that calculates the matrix obtained from 
                                                reduction of order and transforms it to be directly 
                                                applied for Euler Methods
    Output: s (np.array of float-containing arrays) -> Array containing the time evolution of the State Vector
    Author: R.S.S.G.
    Date created: 10/03/2025
    """

    # Unpack necessary parameters from dictionary
    t_span = pdic['t_span']
    dt = pdic['dt']
    s0 = np.array([pdic['x0'], pdic['y0'], pdic['z0']])
    # Initialize time array
    t_axis = np.arange(t_span[0], t_span[1] + dt, dt)
    n = len(t_axis)
    
    # Initialize solution array and set initial condition
    s = np.zeros((n, len(s0)))
    s[0] = s0
    
    # Time integration
    for j in range(0, n-1):
        # Obtain the proper matrix for this method
        m_t = red_ord_matrix(f_slope_matrix(pdic, t_axis[j],s[j])[0], dt,"t")
        
        s[j+1] = np.dot(m_t, s[j])
        
    return s
```

**(c)** Design your own third-order RK method (RK3), and create a python function for this integrator.


As we want a third order method, we need to use 3 slopes such that
$$
\begin{eqnarray*}
k_1 & = & F(t_j,S(t_j))\\
k_2 & = & F\left(t_j+p_1 \, h,S(t_j)+ q_{11} \,k_1\, h\right)\\
k_3 & = & F\left(t_j+p_2\, h,S(t_j)+q_{21}\,k_1\,h+q_{22}\,k_2\,h\right)
\end{eqnarray*}
$$

The solution is then constructed with the step size $h$ as follows:
$$
S(t_{j+1}) = S(t_j) + h\left(c_1\,k_1 + c_2\,k_2 + c_3\,k_3\right).
$$

To obtain the parameters $c_i$, $p_i$, $q_{ij}$, first we get the Taylor series expansion of $S\left(t_n + h \right)$ about $t_n$ up to $O(h^4)$. Then, we need to adjust the solution constructed with our proposed RK3 method to compare it with the Taylor expansion, this step requires expand the slope function and group terms in a similar way to the Taylor expansion.
(To see the whole process and expansions, we recommend the following resource [1] )

[1] Numerical Analysis/Order of RK methods/Derivation of a third order RK method - Wikiversity. (n.d.). Retrieved March 11, 2025, from https://en.wikiversity.org/wiki/Numerical_Analysis/Order_of_RK_methods/Derivation_of_a_third_order_RK_method

After comparison, we obtain the following conditions:
$$
\begin{aligned}
& c_1+c_2+c_3=1 \\
& p_1 c_2+p_2 a_3=1 / 2 \\
& q_{11} c_2+c_3\left(q_{21}+q_{22}\right)=1 / 2 \\
& c_2 p_1^2+c_3 p_2^2=1 / 3 \\
& c_2 p_1 q_{11}+c_3 p_2\left(q_{21}+q_{22}\right)=1 / 3 \\
& c_3 q_{22} p_1=1 / 6 \\
& c_2 q_{11}^2+c_3\left(q_{21}+q_{22}\right)^2=1 / 3 \\
& c_3 q_{22} q_{11}=1 / 6
\end{aligned}
$$

Now, in this system, two equations can be obtained from the six, therefore we need to fix (choose) the values for two of the parameters to obtain the others! 
A common choice is $p_2 = 1, q_{11} = \frac{1}{2}$, thus 
$$
\begin{aligned}
c_1 = \frac{1}{6} \\
c_2 = \frac{2}{3} \\
c_3 = \frac{1}{6} \\
p_1 = \frac{1}{2} \\
q_{11} = \frac{1}{2} \\
q_{21} = -1 \\
q_{22} = 2
\end{aligned}
$$

Finally, our method is given by
$$
\begin{eqnarray*}
k_1 & = & F(t_j,S(t_j))\\
k_2 & = & F\left(t_j+\frac{1}{2} \, h,S(t_j)+ \frac{1}{2}\,k_1 h\right)\\
k_3 & = & F\left(t_j+h,S(t_j)-k_1\,h+2\,k_2\,h\right)
\end{eqnarray*}
$$

The solution is then constructed with the step size $h$ as follows:
$$
S(t_{j+1}) = S(t_j) + \frac{h}{6}\left(k_1 + 4k_2 + k_3\right).
$$

```python
def RK_3(f,pdic):
    """
    Implements the Third-Order Runge-Kutta (RK3) method for ODE integration.
    Input:
        f (python function) -> Function computing the slope F(S). It must be in the form `f(pdic, t, s)`
        pdic (dict) -> dictionary containing the parameters in the system, i.e, g1, g2, c1, 
                            c2, p1, p2, e1, e2, d, x0, y0, z0, t_span, dt, method.
    Output:
        s (np.array of float-containing arrays) -> Array containing the time evolution of the State Vector.
    Author: R.S.S.G.
    Date created: 10/03/2025
    """
    # Unpack necessary parameters from dictionary
    t_span = pdic['t_span']
    dt = pdic['dt']
    s0 = np.array([pdic['x0'], pdic['y0'], pdic['z0']])
    # Initialize time array
    t_axis = np.arange(t_span[0], t_span[1] + dt, dt)
    n = len(t_axis)
    
    # Initialize solution array and set initial condition
    s = np.zeros((n, len(s0)))
    s[0] = s0
    
    # Time integration
    for j in range(0, n - 1):
        # Defining slopes
        k1 = f(pdic, t_axis[j],s[j])
        k2 = f(pdic, t_axis[j] + dt/2,s[j] + dt*k1/2)
        k3 = f(pdic, t_axis[j]+dt, s[j] - dt*k1 + 2*dt*k2)
        # Store solution        
        s[j+1] = s[j] + (dt/6)*(k1 + 4*k2 + k3)
    return s
```

**(d)** Write a python function for the **Butcher's Runge-Kutta method**, which is a popular method for integrating ODEs with a high order of accuracy. It is obtained by following a similar approach to the one we discussed in class. This method uses six points $k_1, k_2, k_3$, $k_4$, $k_5$, and $k_6$. A weighted average of these points is used to produce the approximation of the solution. The algorithm relies on computing the following slopes:

$$
\begin{eqnarray*}
k_1 & = & F(t_j,S(t_j))\\
k_2 & = & F\left(t_j+\frac{h}{4},S(t_j)+\frac{1}{4}k_1h\right)\\
k_3 & = & F\left(t_j+\frac{h}{4},S(t_j)+\frac{1}{8}k_1h+\frac{1}{8}k_2h\right)\\
k_4 & = & F\left(t_j+\frac{h}{2},S(t_j)-\frac{1}{2}k_2h+k_3h\right)\\
k_5 & = & F\left(t_j+\frac{3h}{4},S(t_j)+\frac{3}{16}k_1h+\frac{9}{16}k_4h\right)\\
k_6 & = & F\left(t_j+h,S(t_j)-\frac{3}{7}k_1h+\frac{2}{7}k_2h+\frac{12}{7}k_3h-\frac{12}{7}k_4h+\frac{8}{7}k_5h\right)
\end{eqnarray*}
$$

The solution is then constructed with the step size ($h$, same as `dt`)as follows:

$$
S(t_{j+1}) = S(t_j) + \frac{h}{90}\left(7k_1 + 32k_3 + 12k_4 + 32k_5+7k_6\right).
$$


```python
def Butcher_RK(f, pdic):
    """
    Function implementing the Butcher's Runge-Kutta method for ODE integration.
    Input:
        f (python function) -> Function computing the slope F(S). It must be in the form `f(pdic, t, s)`
        pdic (dict) -> dictionary containing the parameters in the system, i.e, g1, g2, c1, 
                            c2, p1, p2, e1, e2, d, x0, y0, z0, t_span, dt, method.
    Output:
        s (np.array of float-containing arrays) -> Array containing the time evolution of the State Vector.
    Author: R.S.S.G.
    Date created: 10/03/2025
    """
    # Unpack necessary parameters from dictionary
    t_span = pdic['t_span']
    dt = pdic['dt']
    s0 = np.array([pdic['x0'], pdic['y0'], pdic['z0']])
    
    # Initialize time array
    t_axis = np.arange(t_span[0], t_span[1] + dt, dt)
    n = len(t_axis)
    
    # Initialize solution array and set initial condition
    s = np.zeros((n, len(s0)))
    s[0] = s0
    
    # Time integration using Butcher's Runge-Kutta method
    for j in range(n - 1):
        # Compute slopes
        k1 = f(pdic, t_axis[j],s[j])
        k2 = f(pdic, t_axis[j] + dt / 4, s[j] + (dt / 4) * k1)
        k3 = f(pdic, t_axis[j] + dt / 4, s[j] + (dt / 8) * k1 + (dt / 8) * k2)
        k4 = f(pdic, t_axis[j] + dt / 2, s[j] - (dt / 2) * k2 + dt * k3)
        k5 = f(pdic, t_axis[j] + 3 * dt / 4, s[j] + (3 * dt / 16) * k1 + (9 * dt / 16) * k4)
        k6 = f(pdic, t_axis[j] + dt, s[j] - (3 * dt / 7) * k1 + (2 * dt / 7) * k2 \
                               + (12 * dt / 7) * k3- (12 * dt / 7) * k4 + (8 * dt / 7) * k5)
        # Store solution
        s[j + 1] = s[j] + (dt / 90) * (7 * k1 + 32 * k3 + 12 * k4 + 32 * k5 + 7 * k6)
    
    return s
```

### Time stepping:

**(f)** Create a python function that integrates the system of ODEs using the settings and initial conditions as arguments. The function should perform integration for all three methods (according to the option `method`) and return the resulting times ($t$) and system states $S(t)$ as arrays.

```python
def integrators(pdic):
    """
    Wrapper function to select and apply the appropriate ODE integration method.
    Input:
        pdic (dict) -> Dictionary containing the parameters of the system, including:
                       - method: String specifying the integration method ('RK3', 'Trpz', or 'Butcher_RK'). Default-> Butcher RK
                       - Other necessary parameters for the selected method, i.e, g1, g2, c1, 
                            c2, p1, p2, e1, e2, d, x0, y0, z0, t_span, dt.
    Output:
        t (np.array of float) -> Array containing the time series.
        s (np.array of float-containing arrays) -> Array containing the time evolution of the State Vector.

    Author: R.S.S.G.
    Date created: 10/03/2025
    """
    # Unpack necessary parameters from dictionary
    method = pdic['method']
    t_span = pdic['t_span']
    dt = pdic['dt']
    t_axis = np.arange(t_span[0], t_span[1] + dt, dt)
    
    if method == 'RK3':
        s = RK_3(f_slope, pdic)
    elif method == 'Trpz':
        s = trpz(f_slope, pdic)
    else:
        s = Butcher_RK(f_slope, pdic)
    return t_axis, s    
```

```python
class Population_dynamics:
    """
    Class representing the population dynamics given an initial settings passed as a dictionary containing the parameters 
    in the system, i.e, g1, g2, c1, c2, p1, p2, e1, e2, d, x0, y0, z0, t_span, dt, method.
    It requires numpy, matplotlib, scipy. 
    """
    def __init__(self, settings):
        """
        Initialize the class using the settings dictionary.
        Input: settings (dictionary) -> Initial conditions and parameters
        """
        if isinstance(settings, dict):
            self.settings = settings
        else:
            print("Provide initial settings in a proper format, i.e, dictionary")
            raise NotImplementedError

    def f_slope(self, t, s):
        """
        Method representing the slope F(S) of the given system of ODEs 
        Input:self
            t (float) -> Time instance at which we are computing the slope
            s (np.array)-> State vector s.t s[0]=x, s[1]=y, s[2]=z.
        Output: slope (np.array)-> Slope vector function   
        Author: R.S.S.G.
        Date created: 10/03/2025
        """
        # Get state variables
        x, y, z = s[0], s[1], s[2]
        
        # Obtain the dictionary with all the settings
        pdic = self.settings
        # Unpack parameters
        g1, g2, c1, c2, p1, p2, e1, e2, d = pdic['g1'], pdic['g2'], pdic['c1'], pdic['c2'],\
                                            pdic['p1'], pdic['p2'], pdic['e1'], pdic['e2'],\
                                            pdic['d']
        # Getting the slope
        slope = np.array([g1*x*(1-(x/c1))-p1*x*z,
                          g2*y*(1-(y/c2))-p2*y*z,
                          e1*p1*x*z + e2*p2*y*z - d*z])
        return slope

    def trpz(self):
        """
        Function implementing the Trapezoidal Euler method for ODE integration. 
        It uses Euler explicit for the predictor step.
        Input: self
        Output: (np.array of float-containing arrays) -> Array containing the time evolution of the State Vector
        Author: R.S.S.G.
        Date created: 10/03/2025
        """
        # Obtain the dictionary with all the settings
        pdic = self.settings
        f = self.f_slope
        
        # Unpack necessary parameters from dictionary
        t_span = pdic['t_span']
        dt = pdic['dt']
        s0 = np.array([pdic['x0'], pdic['y0'], pdic['z0']])
        
        # Initialize time array
        t_axis = np.arange(t_span[0], t_span[1] + dt, dt)
        n = len(t_axis)
        
        # Initialize solution array and set initial condition
        s = np.zeros((n, len(s0)))
        s[0] = s0
        
        # Time integration
        for j in range(n - 1):
            # Predictor step: Explicit Euler
            s_pred = s[j] + dt * f(t_axis[j], s[j])
            # Corrector step: Trapezoidal method
            s_corr = s[j] + (dt / 2) * (f(t_axis[j], s[j]) + f(t_axis[j], s_pred))
            # Store the computed state
            s[j+1] = s_corr
        return s

    def RK_3(self):
        """
        Implements the Third-Order Runge-Kutta (RK3) method for ODE integration.
        Input: self
        Output: s (np.array of float-containing arrays) -> Array containing the time evolution of the State Vector.
        Author: R.S.S.G.
        Date created: 10/03/2025
        """
        # Obtain the dictionary with all the settings
        pdic = self.settings
        f = self.f_slope
        # Unpack necessary parameters from dictionary
        t_span = pdic['t_span']
        dt = pdic['dt']
        s0 = np.array([pdic['x0'], pdic['y0'], pdic['z0']])
        # Initialize time array
        t_axis = np.arange(t_span[0], t_span[1] + dt, dt)
        n = len(t_axis)
        
        # Initialize solution array and set initial condition
        s = np.zeros((n, len(s0)))
        s[0] = s0
        
        # Time integration
        for j in range(n - 1):
            # Defining slopes
            k1 = f(t_axis[j], s[j])
            k2 = f(t_axis[j] + dt/2, s[j] + dt*k1/2)
            k3 = f(t_axis[j] + dt, s[j] - dt*k1 + 2*dt*k2)
            # Store solution        
            s[j+1] = s[j] + (dt/6)*(k1 + 4*k2 + k3)
        return s

    def Butcher_RK(self):
        """
        Function implementing the Butcher's Runge-Kutta method for ODE integration.
        Input: self
        Output: s (np.array of float-containing arrays) -> Array containing the time evolution of the State Vector.
        Author: R.S.S.G.
        Date created: 10/03/2025
        """
        # Obtain the dictionary with all the settings
        pdic = self.settings
        f = self.f_slope
        
        # Unpack necessary parameters from dictionary
        t_span = pdic['t_span']
        dt = pdic['dt']
        s0 = np.array([pdic['x0'], pdic['y0'], pdic['z0']])
        
        # Initialize time array
        t_axis = np.arange(t_span[0], t_span[1] + dt, dt)
        n = len(t_axis)
        
        # Initialize solution array and set initial condition
        s = np.zeros((n, len(s0)))
        s[0] = s0
        
        # Time integration using Butcher's Runge-Kutta method
        for j in range(n - 1):
            # Compute slopes
            k1 = f(t_axis[j], s[j])
            k2 = f(t_axis[j] + dt / 4, s[j] + (dt / 4) * k1)
            k3 = f(t_axis[j] + dt / 4, s[j] + (dt / 8) * k1 + (dt / 8) * k2)
            k4 = f(t_axis[j] + dt / 2, s[j] - (dt / 2) * k2 + dt * k3)
            k5 = f(t_axis[j] + 3 * dt / 4, s[j] + (3 * dt / 16) * k1 + (9 * dt / 16) * k4)
            k6 = f(t_axis[j] + dt, s[j] - (3 * dt / 7) * k1 + (2 * dt / 7) * k2 + (12 * dt / 7) * k3 \
                  - (12 * dt / 7) * k4 + (8 * dt / 7) * k5)
            
            # Store solution
            s[j + 1] = s[j] + (dt / 90) * (7 * k1 + 32 * k3 + 12 * k4 + 32 * k5 + 7 * k6)
        
        return s

    def integrators(self):
        """
        Wrapper function to select and apply the appropriate ODE integration method.
        Input: self
        Output: t (np.array of float) -> Array containing the time series.
                s (np.array of float containing arrays) -> Array containing the time evolution of the State Vector.
        Author: R.S.S.G.
        Date created: 10/03/2025
        """
        pdic = self.settings
        method = pdic['method']
        t_span = pdic['t_span']
        dt = pdic['dt']
        t_axis = np.arange(t_span[0], t_span[1] + dt, dt)
        
        if method == 'RK3':
            s = self.RK_3()
        elif method == 'Trpz':
            s = self.trpz()
        else:
            s = self.Butcher_RK()
        return t_axis, s    

    def plotting_integrator(self):
        """
        Generates a time-series plot of the state vector components using the selected integration method.
        Input: self
        Output:fig (matplotlib.figure.Figure) -> A figure containing three subplots for the evolution of x(t), y(t), and z(t).
        Author: R.S.S.G.
        Date created: 10/03/2025
        """
        # Define time and s vector arrays, method and time step.
        t_axis, s, method, dt = *self.integrators(), self.settings['method'], self.settings['dt']

        # Identify the method
        if method == "Trpz":
            m = "Trapezoidal"
        elif method == "RK3":
            m = method
        else:
            m = "Butcher_RK"
        
        # Extract results
        x_vals, y_vals, z_vals = s[:, 0], s[:, 1], s[:, 2]
                
        # Determine the global y-axis range, motivated by the carrying capacities.
        y_min = -5
        y_max = 210
        
        # Create the figure with three subplots
        fig, ax = plt.subplots(1, 3, figsize=(17, 6), sharey=True)
        
        # Plot each variable
        ax[0].plot(t_axis, x_vals, label=r'$x(t)$ (Prey 1)', color='blue')
        ax[0].plot(t_axis[0], x_vals[0], label=r"Initial Conditions $x(0)$", marker = "d", color = "k")
        ax[1].plot(t_axis, y_vals, label=r'$y(t)$ (Prey 2)', color='green')
        ax[1].plot(t_axis[0], y_vals[0], label=r"Initial Conditions $y(0)$", marker = "d", color = "k")
        ax[2].plot(t_axis, z_vals, label=r'$z(t)$ (Predator)', color='red')
        ax[2].plot(t_axis[0], z_vals[0], label=r"Initial Conditions $z(0)$", marker = "d", color = "k")
        
        for a in ax:
            a.set_xlabel('Time')
            a.set_ylim(y_min, y_max)  # Apply the same y-axis range
            a.legend()
            a.grid(True, linestyle='--', alpha=0.7)
        
        # Set common y-axis label
        ax[0].set_ylabel('Population')
        
        plt.suptitle(f'Population Dynamics using {m} method')
        plt.tight_layout()
        plt.close()
        return fig

    def error(self):
        """
        Computes the error L2 norm between the computed numerical method solution approximation and the reference state vector from scipy.integrate.solve_ivp()
        using DOP853 method.
        It requires scipy.integrate.solve_ivp and numpy.linalg
        Input: self
        Output: error (float) -> L2 Norm of the difference between the computed solution and the reference state vector.
                t_axis (np.array of floats) -> Time series.
                s_ref (np.array of float containing arrays) -> Array containing the time evolution of the reference state vector from scipy
        Author: R.S.S.G.
        Date created: 10/03/2025
        """
        
        t_axis, s = self.integrators()
        
        # Compute solution:
        sln = solve_ivp(self.f_slope,[t_axis[0],t_axis[-1]], np.array([self.settings['x0'],self.settings['y0'],self.settings['z0']]),\
                        method= "DOP853", t_eval = t_axis,atol = 10**-13,rtol=10**-13) 
        s_ref = sln.y.T
        
        if isinstance(s_ref, np.ndarray):
            try:
                error = la.norm(s - s_ref)
            except:
                print(f"Check array shapes or transpose {sln}")
                
            return error, t_axis, s_ref
        else:
            print(f"Cannot compare {type(s_ref)} with the state vector.")
            raise NotImplementedError
```

### Equilibrium conditions:

**(c)** Write down the equilibrium condition for the system, $S(t)$. Then, create a python function that uses sympy to study the equilibrium populations of $x$, $y$, and $z$ for a range of values of the natural death rate of the predator, $d$. The function should accept a range of $d$ values and return the equilibrium populations for all the $d$ values in the range. Ensure that the solutions are filtered to exclude extinction cases.


As mentioned previously, the equilibrium condition is given by $\vec{F \left( \vec{S} \right)}=\vec{0}$ 
$$\vec{F \left( \vec{S} \right) }=\left(\begin{array}{l}
g_1\,x \left( 1 - \frac{x}{c_1} \right) - p_1\,x\,z, \\
g_2\,y \left( 1 - \frac{y}{c_2} \right) - p_2\,y\,z, \\
e_1\,p_1\,x\,z + e_2\,p_2\,y\,z - d\,z.
\end{array}\right) = \left(\begin{array}{l}
0\\
0\\
0
\end{array}\right) $$

```python
def equilibrium_populations(d_values, pdic):
    """
    Computes the equilibrium populations, excluding extinction cases, for a predator-prey system given a range of death rates.
    It requires sympy & numpy
    Input:  d_values (array of floats) -> Array of death rate values.
            pdic (dict) -> dictionary containing the parameters in the system, i.e, g1, g2, c1, 
                            c2, p1, p2, e1, e2, d, x0, y0, z0, t_span, dt, method.
    Output:
        solutions (np.array of shape (len(d_values), 3)) -> Array containing equilibrium populations:
                                                            - Column 0: Equilibrium values for prey population 1 (x).
                                                            - Column 1: Equilibrium values for prey population 2 (y).
                                                            - Column 2: Equilibrium values for predator population (z).
                                                            - If no valid equilibrium exists (extinction case) for a given `d_value`, the corresponding row contains nan.
    Author: R.S.S.G.
    Date created: 13/03/2025
    """
    # Unpack parameters
    g1, g2, c1, c2, p1, p2, e1, e2 = pdic['g1'], pdic['g2'],pdic['c1'],pdic['c2'],\
                                        pdic['p1'], pdic['p2'], pdic['e1'], pdic['e2']
    
    
    x, y, z, d = sp.symbols('x y z d', real=True, positive=True)
    
    # Define the system of equations
    eq1 = g1 * x * (1 - (x / c1)) - p1 * x * z
    eq2 = g2 * y * (1 - (y / c2)) - p2 * y * z
    eq3 = e1 * p1 * x * z + e2 * p2 * y * z - d * z
    
    # Solve for equilibrium points
    solutions = sp.solve([eq1, eq2, eq3], (x, y, z), dict=True)
    results = []
    for d_val in d_values:
        valid_solutions = []
        for sol in solutions:
            # Substitute d with the given value
            sol_evaluated = [sol[var].subs(d, d_val) for var in sol]
            
            # Check if all populations are positive to ensure non extinction cases
            if all(sol.is_real and sol > 0 for sol in sol_evaluated):
                valid_solutions.append(sol_evaluated)
        results.append(valid_solutions)
        
    solutions = np.empty((len(ds),3))
    solutions[:] = np.nan
    j = 0
    for i in results:
        if len(i) ==1 :
            solutions[j] = i[0]
        j += 1
    return solutions
```

To obtain the exact values for the interval, we can use simpy. As the only non-extinction case solution is the last one printed above, we can solve for each variable so that it's bigger than zero.

```python
def solv_an():
    # Define the symbols
    x, y, z = sp.symbols('x y z')
    g1, g2, c1, c2, p1, p2, e1, e2, d = sp.symbols('g1 g2 c1 c2 p1 p2 e1 e2 d')
    
    # Define the equations
    eq1 = g1 * x * (1 - x / c1) - p1 * x * z
    eq2 = g2 * y * (1 - y / c2) - p2 * y * z
    eq3 = e1 * p1 * x * z + e2 * p2 * y * z - d * z
    
    # Solve eq1 for x
    x_solutions = sp.solve(eq1, x)
    
    # Solve eq2 for y
    y_solutions = sp.solve(eq2, y)
    
    # Substitute x and y solutions into eq3 and solve for z
    solutions = []
    for x_sol in x_solutions:
        for y_sol in y_solutions:
            eq3_sub = eq3.subs({x: x_sol, y: y_sol})
            z_solutions = sp.solve(eq3_sub, z)
            for z_sol in z_solutions:
                solutions.append({x: x_sol, y: y_sol, z: z_sol})
    return solutions

# Calling the function
solv_an()
```

```python
def exactds(pdic):
        
    # Unpack parameters
    g1, g2, c1, c2, p1, p2, e1, e2 = pdic['g1'], pdic['g2'],pdic['c1'],pdic['c2'],\
                                        pdic['p1'], pdic['p2'], pdic['e1'], pdic['e2']
    z, d = sp.symbols('z d')
    # Define the equation for z as the other variables are written as explicit functions of z
    eq0 = z-(g1*g2*(c1*e1*p1 + c2*e2*p2 - d)/(c1*e1*g2*p1**2 + c2*e2*g1*p2**2))
    # Obtaining z in terms of d
    sol0 = sp.solve(eq0, z)[0]
    # Solving for d such that z>0
    sol1 = sp.solve(sol0, d)[0]
    # Solving for d such that x and y>0
    eq1 = c1*(g1 - p1*(sol0))/g1
    eq2 = c2*(g2 - p2*(sol0))/g2
    sol2 = sp.solve(eq1, d)[0]
    sol3 = sp.solve(eq2, d)[0]
    sols = [sol1,sol2,sol3]
    sols.sort()
    # Now, we will return only d positive values
    return [i for i in sols if i > 0]

# Calling our function
d_valex = exactds(pdic)
print(f"The numerically obtained d interval for allowed equilibrium populations is: {d_valex}")
```

## 3. Quantum Harmonic Oscillator (7 points)

The SchrÃ¶dinger equation for the quantum harmonic oscillator is:

$$-\frac{\hbar^2}{2m}\frac{d^2\psi}{dx^2}+\frac{1}{2}m\omega^2x^2\psi=E\psi$$

It can be rewritten, in terms of a new variable, $\xi\equiv\sqrt{\frac{m\,\omega}{\hbar}}x$, as follows:

$$\frac{d^2\psi}{d\xi^2}=\left(\xi^2 -K\right)\psi$$

where $K\equiv\frac{2E}{\hbar\omega}$ is the energy in units of $\frac{1}{2}\hbar\omega$.

### Order reduction and slope function:

**(a)** Reduce the above ODE to first order. Write down the resulting slope function.


Now, let's define the state vector $\vec{S}$ as

$$
\vec{S} =\left(\begin{array}{c}
\psi\\
\frac{d\psi}{d\xi}
\end{array}\right)
$$

Then,
$$
\frac{d\vec{S}}{d\xi} = \left(\begin{array}{c}
\frac{d\psi}{d\xi} \\
(\xi^2 - K)\, \psi
\end{array}\right) = \left(\begin{array}{c}
 0 & 1\\
\xi^2 - K & 0
\end{array}\right) \vec{S} = \vec{F}
$$

Notice that, as expected from the linear behaviour of QM, the slope can be written explicitely as function of the sate vector $\vec{S}$

```python
# Let's define the slope in a function
def qmslope(xi, s, k):
    """
    Computes the slope of the quantum harmonic oscillator ODE based on the given parameters.
    It requires numpy.
    Input:  xi (float or int) -> New variable to simplify the ODE, s.t, psi(xi) is the wavefunction.
           s (array-like) -> A vector representing the state of the system [psi, dpsi/dxi].
           k (float or int) -> A parameter related to the system's energy.
    Output: slope (np.array) -> Vector slope.
    Author: R.S.S.G.
    Date created: 13/03/2025
    """
    m = np.asarray([[0,1],[xi**2 - k, 0]], dtype = object)
    slope = np.asarray(np.dot(m, s), dtype = object)
    return slope
```

### Shooting method class:

**(b)** Carefully read the tasks (c-g) below and design a suitable python class with attributes and methods that solves the above ODE using **the shooting method** with the slope computed in (a). You may use scipy integrators; there is no need to design your own integrators in this problem.


```python
class QHO:
    """
    Class to solve the Quantum Harmonic Oscillator (QHO) problem.
    It requires numpy, scipy, matplotlib.
    """
    def __init__(self, settings):
        """
        Initialization method, requires initial conditions and parameter K to be 
        passed as np.array s.t settings = np.array([psi(0), dpsi/dxi (0), K])
        Input: settings (np.array) -> Settings described above.
        
        Author: R.S.S.G.
        Date created: 13/03/2025
        """
        if isinstance(settings, np.ndarray) and settings.shape == (3,):
            self.settings = settings  # Define settings attribute
        else:
            print("Provide initial conditions in a proper numpy array of shape (3,).")
            raise ValueError("Invalid settings.")

    @staticmethod
    def qho_slope(xi, s, k):
        """
        Computes the slope of the quantum harmonic oscillator ODE based on the given parameters.
        It requires numpy.
        Input:  xi (float or int) -> New variable to simplify the ODE, s.t, psi(xi) is the wavefunction.
               s (array-like) -> A vector representing the state of the system [psi, dpsi/dxi].
               k (float or int) -> A parameter related to the system's energy.
        Output: slope (np.array) -> Vector slope.
        Author: R.S.S.G.
        Date created: 13/03/2025
        """
        m = np.asarray([[0, 1], [xi**2 - k, 0]], dtype = object)
        slope = np.dot(m, s)
        return slope

    def solve_qho(self, k=None, ics=None):
        """
        Function solving the system using initial conditions and DOP853 as numerical integration method.
        It requires numpy and scipy.integrate.solve_ivp.
        Input: self
                k (float) -> K parameter
                ics (np.array of floats) -> Initial conditions of the wavefunction given as [psi(0), dpsi/dxi (0)]
        Output:
            sol (scipy.integrate.OdeResult) -> Scipy object containing the solution to the system:
                                               - sol.t: Array of time points (xi values).
                                               - sol.y: Array of solution values (psi and dpsi/dxi).
        Author: R.S.S.G.
        Date created: 13/03/2025
        """
        if ics is None:
            ics = np.array([self.settings[0], self.settings[1]])
        if k is None:
            k = self.settings[2]
        # Define step size and interval for the solution
        h_step = 0.00001
        
        # We've defined this interval by trial and error
        xi_span = [-20, 20 + h_step]
        
        # Define xi axis
        xi_axis = np.arange(xi_span[0], xi_span[1], h_step)
        
        # Integrating using scipy
        sol = solve_ivp(lambda xi, s: self.qho_slope(xi, s, k), t_span=xi_span, 
                        y0=ics, t_eval=xi_axis, method="DOP853", rtol=10e-5)
        
        return sol

    def obj_qho(self, k, ics):
        """
        Objective function that will be used in optimization routine using fsolve.
        Input:  
            k (float or int) -> Parameter related to the system's energy. 
        Output:
            final_psi (float) -> The value of the wavefunction (psi) at the final point of the solution
                                to be compared with zero, as psi should vanish for large xi.
        Author: R.S.S.G.
        Date created: 13/03/2025
        """
        # Using the functional psi = 0 for large xi
        sol = self.solve_qho(k=k, ics=ics)
        return sol.y[0][-1]
        
    def e_level(self, n, v=True):
        """
        Function that calculates the K paramenter and energy of the nth level using shooting method.
        It requires scipy.
        Input:  self
                n (int) -> Number of the wanted level. Ground state is defined as n=0.
                v (Boolean, optional) -> Verbose, if True prints the process and result of K parameter. Otherwise
                                it will just return the value. Default = True
        Output:
            k_g (float) -> K parameter of the nth level.
        Author: R.S.S.G.
        Date created: 13/03/2025
        """
        if n%2 == 0: # Check for even states
            # Define initial conditions for even states
            ics = np.array([self.settings[0], self.settings[1]])    
        else: # For odd states
            # Setting the initial conditions for odd states
            ics = np.array([self.settings[0], 1.])

        # Computing optimal k for nth level state
        k_g, = fsolve(self.obj_qho, x0 = (2*n + 0.99), args=(ics,), xtol = 1e-8) # Here we just vary K parameter taking into account that the k parameter for the levels should be odd numbers
        if v:
            print(f"The {n}th level K parameter, numerically, is: K = ", k_g)
            print(f"Truncating to 6 significant figures, K_{n} = {round(k_g):.6f}  ")
            print(f"The energy, numerically, is: E_{n} = {round(k_g)}(hbar * omega)/2 ")
        
        k_g = round(k_g)
        
        return k_g

    def plotting_ks(self, ks, colors, n):
        """
        Function that plots the wavefunctions for a given set of k values in a multi-panel figure.
        It requires matplotlib.
        Input:  self
                ks (np.array) -> Array of k values corresponding to the energy levels.
                colors (list) -> List of colors for plotting each wavefunction.
                n (int) -> Number of the wanted level. Ground state is defined as n=0.
        Output:
            fig (matplotlib.figure.Figure) -> Figure object containing the multi-panel plot.
        Author: R.S.S.G.
        Date created: 13/03/2025
        """
        
        if n%2 == 0: # Check for even states
            # Define initial conditions for even states
            ics = np.array([self.settings[0], self.settings[1]])    
        else: # For odd states
            # Setting the initial conditions for odd states
            ics = np.array([self.settings[0], 1.])
        
    
        # Let's obtain the xi interval and solutions
        xi_val = self.solve_qho(ks[0], ics).t
        sols = [self.solve_qho(k, ics).y[0] for k in ks]
        # Let's scale the wavefunctions to have a good representation
        sols = [sol / sol[:len(sol) // 2].max() for sol in sols]
        
        # Create the multi-panel figure using GridSpec
        fig = plt.figure(figsize=(14, 10))
        gs = fig.add_gridspec(2, 3)  # 2 rows, 3 columns

        fig.suptitle(fr"{n}th State Wave Function $\psi$ vs $\xi$ (varying K)")
        # Top 3 subplots
        ax1 = fig.add_subplot(gs[0, 0])
        ax2 = fig.add_subplot(gs[0, 1])
        ax3 = fig.add_subplot(gs[0, 2])
        middle = round((len(ks)+0.1)/2)
        # Bottom subplot (spanning all 3 columns)
        ax_bottom = fig.add_subplot(gs[1, :])  # This spans the entire bottom row
        
        # Plot the top 3 subplots
        for i, ax in enumerate([ax1, ax2, ax3]):
            # Select every (len(ks) // 3-th solution for the top subplots
            idx = i * (len(ks) // 3)
            
            ax.plot(xi_val, sols[idx], label=f"k = {ks[idx]:.4f}", color=colors[idx])
            ax.set_xlabel(r'$\xi$')
            ax.set_ylabel(r'$\psi$')
            ax.set_ylim(-max(sols[middle][:len(sols[middle])//2]), 1.4 * max(sols[middle][:len(sols[middle])//2]))  # Adjust ylim based on the largest solution
            ax.set_xlim(-7.5, 7.5)  # Adjust xlim as needed
            ax.legend()
        # Plot the bottom subplot (combined plot)
        for j, sol in enumerate(sols):
            ax_bottom.plot(xi_val, sol, label=f"k = {ks[j]:.3f}", color=colors[j])
        
        ax_bottom.set_xlabel(r'$\xi$')
        ax_bottom.set_ylabel(r'$\psi$')
        ax_bottom.set_ylim(-max(sols[middle][:len(sols[middle])//2]), 1.4 * max(sols[middle][:len(sols[middle])//2]))  # Adjust ylim based on the largest solution
        ax_bottom.set_xlim(-7.5, 7.5)  # Adjust xlim as needed
        ax_bottom.legend()
        # Adjust layout and show the plot
        plt.tight_layout()
        plt.show()
        plt.close()
        return fig

    def plot_v_psi(self, hbar, omega, n_states):
        """
        Function that plots the harmonic potential, energy levels, and normalized wavefunctions for the first n_states.
        It requires matplotlib.
        Input:  self
                hbar (float) -> Reduced Planck's constant.
                omega (float) -> Angular frequency of the harmonic oscillator.
                n_states (int) -> Number of energy levels to plot.
        Output:
            fig (matplotlib.figure.Figure) -> Figure object containing the plot.
        Author: R.S.S.G.
        Date created: 13/03/2025
        """
        # Let's define the harmonic potential in terms of xi
        harmonic_potential = lambda xi, hbar, omega: 0.5*hbar*omega*(xi**2)
        
        # Computing the energy levels for n = 0, 1, 2, ..., n_states-1
        ks = np.array([self.e_level(n, v=False) for n in range(n_states)])
        es = ks / 2  # Energy levels
        
        # Solve for the wavefunctions
        xi_vals = self.solve_qho(ks[0]).t
        sols = [self.solve_qho(k).y[0] for k in ks]
        
        # Let's scale the wavefunctions to have a good representation
        # Notice that this is not normalizing them.
        sols = [sol / sol[:len(sol) // 2].max() for sol in sols]
        
        # Plot the harmonic potential, energy levels, and wave functions
        xis = np.linspace(-5, 5, 100)
        v_xi = harmonic_potential(xis, hbar, omega)
        
        fig = plt.figure(figsize=(14, 8))
        
        # Plot the harmonic potential
        plt.plot(xis, v_xi, label=r"Harmonic Potential $V(\xi) = \frac{1}{2}\xi^2$", color="black")
        
        # Plot the energy levels and wave functions
        for n in range(n_states):
            # Energy level
            plt.axhline(es[n], color="gray", linestyle='--', alpha=0.8)
            
            # Wavefunction (shifted vertically by the energy level)
            plt.plot(xi_vals, sols[n] + es[n], label=fr"$n = {n}$, $E_{n}={es[n]}$")
        
        # Add labels and legend
        plt.xlabel(r'$\xi$')
        plt.ylabel(r'Energy / Wavefunction $\psi$')
        plt.title('Harmonic Potential, Energy Levels, and Wave Functions')
        plt.legend()
        plt.ylim(-0.5, 8)  
        plt.xlim(-4.5, 4.5) 
        plt.tight_layout()
        plt.show()
        plt.close()
        return fig
```
